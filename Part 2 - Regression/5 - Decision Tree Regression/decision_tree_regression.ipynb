```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Datasets/Position_Salaries.csv")
x = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x,y)
```

```python
#visualization  in graph 
plt.rcParams["figure.figsize"] = (30,20) #for increasing the size of plot
plt.scatter(x , y , color="red")
plt.plot(x,regressor.predict(x))
plt.show()
```

```python
#to get a better visualization we increase precision of data x
#x_grid will be new test set whose predicted values will be plotted

x_grid = np.arange(min(x),max(x),0.01)     #here we create an array starting from min(x) to max(x) with each value increses with 0.01
x_grid = x_grid.reshape((len(x_grid),1)) 
```

```python
plt.rcParams["figure.figsize"] = (30,20)
plt.scatter(x , y , color="red")
plt.plot(x_grid,regressor.predict(x_grid))
plt.show()
```
