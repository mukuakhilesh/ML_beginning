```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
dataset.keys()                   # output is in form of dictionary i.e. key value pairs
```

```python
dataset['target_names']
```

```python
dataset['feature_names']       #the features in dataset
```

```python
#dataset is divided in ttwo arrays
print(dataset['data'][0:6,:])#printing first 5 rows of data...measurements are stored in column named 'data'
print(dataset['target'])     # output of all inputs...stored in column named 'target'
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(dataset['data'],dataset['target'],test_size=0.25,random_state=0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#visualization of data before application
#we will use a pair plot as there are four columns
#scatter_matrix is used for visualization provided by pandas...so first we convert array to dataframe

iris_dataframe = pd.DataFrame(X_train, columns=dataset.feature_names)
print(iris_dataframe.head())
 #creating a scatter plot
plot = pd.plotting.scatter_matrix(iris_dataframe,c=y_train,figsize=(15,15),marker='o',
                        hist_kwds={'bins':20},alpha=0.75)
```

```python
#we will use KNN classification first

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train , y_train)
```

```python
predicted = knn.predict(X_test)

print(knn.score(X_test,y_test))            #score function returns the accuracy of model...it is available in most sklearn models"

print(predicted==y_test)                   #this compares the two arrays and return true where value are same and 0 where they are different
print(np.mean(predicted==y_test))
```
