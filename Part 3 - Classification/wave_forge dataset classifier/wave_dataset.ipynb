```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
dataset = load_wine()
print('features in dataset : \n{}\n'.format(dataset['feature_names']))
print("classes in target or dependent datasets : \n{}\n".format(dataset['target_names']))
#print(dataset['DESCR'])
X = dataset['data']
y = dataset['target']
print("Shape of independent matrix : {}".format(X.shape))
```

```python
dataframe = pd.DataFrame(X,columns=dataset['feature_names'])
dataframe.head()

plot = pd.plotting.scatter_matrix(dataframe , c=y ,figsize=(100,70), alpha=0.75)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size =0.2,random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
test_accuracy=[]
train_accuracy=[]
for i in range(1,20) :    
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train,y_train)
        test_accuracy.append(knn.score(X_test,y_test))
        train_accuracy.append(knn.score(X_train,y_train))

plt.rcParams["figure.figsize"] = (20,15)
plt.plot(range(1,20) , test_accuracy,marker='.',label="test")
plt.plot(range(1,20) , train_accuracy,label="train")
plt.ylabel("Accuracy")
plt.legend()
```

```python
#this shows that KNN model is very ineffective for this datasets
# we also see that accuracy is highest when neighbors = 1. i.e. a very complex model so KNN is not acceptable here
```
