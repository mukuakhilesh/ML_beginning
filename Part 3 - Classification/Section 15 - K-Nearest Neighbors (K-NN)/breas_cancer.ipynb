```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
print("\n columns in dataset : \n{}".format(dataset['feature_names']))
print("\n classes in output: {}".format(dataset['target_names']))
print("\n shape : {}".format(dataset["data"].shape))
X = dataset["data"]
y = dataset["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
test_accuracy=[]
train_accuracy=[]
for i in range(1,20) :    
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train,y_train)
        test_accuracy.append(knn.score(X_test,y_test))
        train_accuracy.append(knn.score(X_train,y_train))

plt.rcParams["figure.figsize"] = (20,15)
plt.plot(range(1,20) , test_accuracy,marker='.',label="test")
plt.plot(range(1,20) , train_accuracy,label="train")
plt.ylabel("Accuracy")
plt.legend()
```
