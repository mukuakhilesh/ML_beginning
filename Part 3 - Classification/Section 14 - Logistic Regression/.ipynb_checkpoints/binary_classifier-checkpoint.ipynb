```python
import pandas as pd
import numpy as np
import mglearn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Datasets/Social_Network_Ads.csv')
data.head(5)
df = pd.get_dummies(data.iloc[:,1:4])
df.head(5)
```

```python
x = df.iloc[:,0:2].values
y = data.iloc[:,-1].values

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy=True ,with_mean = True )
x=scaler.fit_transform(x)

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python

plt.rcParams["figure.figsize"] = (20,15)
mglearn.plots.plot_2d_separator(classifier,x,fill=True,eps=0.5,alpha=0.45)
mglearn.discrete_scatter(x_train[:,0],x_train[:,1],y_train,markers=['.'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
```
